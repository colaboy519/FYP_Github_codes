# Check N convergence
to use this code: 
1. Run pulse optimization with same optimization algorithm setting and same initial pulses
2. Run the optimization for multiple N values, eg: [3,4,5,6,7,8,9,10]
3. Save the output files and keep note of the file names
3. Copy paste the file names below and run this script

specify: 
- n_ts and evo_time 
- N_values
- alpha

```python
### plot the pulses on the same graph ###
import numpy as np
import matplotlib.pyplot as plt
N_values = [3,4,5,6,7,8,9,10]
alpha = 0.5

files = []
for N in N_values:
    files.append(f'ctrl_amps_initial_coherent_to_cat_state_n_ts100_evo_time1_ptypeZERO_alpha{alpha}_N{N}_fid_err_targ0.0001_max_iter100000_max_wall_time10000.txt')


# Load data from the file into a numpy array
datas = [np.genfromtxt(f, delimiter='\t') for f in files]  

# first control pulse
control1 = plt.figure()
for i, data in enumerate(datas):
    plt.plot(datas[i][:, 0], datas[i][:, 1], label=f"N = {N_values[i]}")
plt.xlabel('time steps')
plt.ylabel('pulse amplitude')
plt.title('control 1')
plt.legend()

# second control pulse
control2 = plt.figure()
for i, data in enumerate(datas):
    plt.plot(datas[i][:, 0], datas[i][:, 2], label=f"N = {N_values[i]}")
plt.xlabel('time steps')   
plt.ylabel('pulse amplitude')
plt.title('control 2')
plt.legend()

plt.show()
```
